```python
# Estrategias
strategies=['Rock','Paper','Scissors','Spock','Lizzard']
```

```python
#Reglas

payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Spock', 'Paper'):(0,1),
        ('Paper','Spock'):(1,0),
        ('Lizzard','Paper'):(1,0),
        ('Paper','Lizzard'):(0,1),
        ('Scissors','Lizzard'):(1,0),
        ('Lizzard','Scissors'):(0,1),
        ('Scissors','Spock'):(0,1),
        ('Spock','Scissors'):(1,0),
        ('Spock','Rock'):(1,0),
        ('Rock','Spock'):(0,1),
        ('Spock','Lizzard'):(0,1),
        ('Lizzard','Spock'):(1,0),
        ('Lizzard','Paper'):(1,0),
        ('Paper','Lizzard'):(0,1),
        ('Lizzard','Rock'):(0,1),
        ('Rock','Lizzard'):(1,0),
        ('Spock','Spock'):(0,0),
        ('Lizzard','Lizzard'):(0,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
```

```python
Players=[{'name':'John','score':0,'strategy':None},
         {'name':'Mary','score':0,'strategy':None},
         {'name':'Paty','score':0,'strategy':None},
         {'name':'Alex','score':0,'strategy':None},
         {'name':'Luis','score':0,'strategy':None},
         {'name':'Bruce','score':0,'strategy':None},
         {'name':'Elvis','score':0,'strategy':None},
         {'name':'Jose','score':0,'strategy':None},
         {'name':'Majo','score':0,'strategy':None},
         {'name':'Narda','score':0,'strategy':None},]
```

```python
from random import choice

#simplest strategy: choose randomly
choice(strategies);
```

```python
Players[0]['strategy']=choice(strategies)
Players[1]['strategy']=choice(strategies)
```

```python
Players[0]['strategy'],Players[1]['strategy']
```

```python
# social result of individual decision
result = payoff[Players[0]['strategy'],Players[1]['strategy']]
result
```

```python
# update agents situation
Players[0]['score']+=result[0]
Players[1]['score']+=result[1]
```

```python
Players
```

```python
import pandas as pd

socialResults=pd.DataFrame((Players[0], Players[1]))
socialResults
```

```python
winnerScore=socialResults.score.max()

#social outcome
socialResults[socialResults.score==winnerScore]
```

```python
import itertools

# pair is a tuple of dicts
for pair in itertools.combinations(Players,2):
    print(pair)
```

```python
import itertools

# each dict
for player1,player2 in itertools.combinations(Players,2):
    print(player1,player2)
```

```python

# several rounds
for aRound in range(100):

    # en each round:
    for player1,player2 in itertools.combinations(Players,2):
        # each chooses strategy
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)

        # result from strategy chosen
        result=payoff[player1['strategy'],player2['strategy']]

        # update scores
        player1['score']+=result[0]
        player2['score']+=result[1]

        print(player1['name'] + " chose " + player1['strategy'] +" - "+player2['name'] + " chose " + player2['strategy'] )

        if payoff[player1['strategy'],player2['strategy']] == (1,0):
            print(player1['name'] + " won")
    
        if payoff[player1['strategy'],player2['strategy']] == (0,1):
            print(player2['name'] + " won")
    
        else:
            print("TIE!")
```
